```python
from transformers import pipeline
import gradio as gr
```

```python
# Use BART or T5 (both free)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize_text(text):
    if len(text.split()) < 30:
        return "Please enter a longer text for meaningful summarization."

    try:
        summary = summarizer(text, max_length=50, min_length=20, do_sample=False)  # Forces 2-3 lines
        return summary[0]['summary_text']
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
interface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(label="Enter Long Text to Summarize", lines=10),
    outputs=gr.Textbox(label="Summarized Text (2-3 Lines)"),
    title="Free AI Summarizer (No API Key Required)",
    description="Enter a long paragraph, and the model will summarize it into 2-3 lines.",
    theme="default"
)

# Launch Gradio App
interface.launch(share=True)
```
