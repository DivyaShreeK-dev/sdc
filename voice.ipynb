```python
from gtts import gTTS
from IPython.display import Audio

# Function to generate speech from text
def generate_voice(text, language='en'):
    # Convert the text to speech
    tts = gTTS(text=text, lang=language, slow=False)

    # Save the generated speech to a file
    tts.save("output.mp3")

    # Play the speech directly in the notebook
    return Audio("output.mp3", autoplay=True)

# Example text to be converted to speech
text = "Hello, this is a voice generated by the Google TTS library in Google Colab!"
generate_voice(text)
```

```python
# Import necessary libraries
import gradio as gr
from gtts import gTTS
from IPython.display import Audio

# Function to generate speech from text
def generate_voice(text, language='en'):
    # Convert the text to speech
    tts = gTTS(text=text, lang=language, slow=False)

    # Save the generated speech to a file
    tts.save("output.mp3")

    # Play the speech directly in the notebook
    return Audio("output.mp3", autoplay=True)

# Create a Gradio interface
gr.Interface(fn=generate_voice,
             inputs=gr.Textbox(label="Enter text to convert to speech", lines=2, placeholder="Type your text here..."),
             outputs="audio",  # Output type is audio
             live=True,  # This makes it respond immediately after input
             title="Text-to-Speech Converter",
             description="Type any text, and it will be converted into speech.") \
    .launch()
```
