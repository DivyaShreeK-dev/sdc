```python
import google.generativeai as genai

# Replace 'YOUR_API_KEY' with your actual Gemini API key
genai.configure(api_key="AIzaSyB205sQLpIyc1fteQICREndFQU-bPXFerA")
```

```python
def summarize_text(text):
    try:
        # Load the Gemini model
        model = genai.GenerativeModel("gemini-pro")

        # Request the model to summarize the input text
        response = model.generate_content(f"Summarize the following text:\n\n{text}")

        # Return the summarized text
        return response.text
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
import gradio as gr

# Create Gradio Interface
interface = gr.Interface(
    fn=summarize_text,  # Function to call
    inputs=gr.Textbox(label="Enter Text to Summarize"),  # User Input
    outputs=gr.Textbox(label="Summarized Text"),  # Model Output
    title="Gemini AI Text Summarization",
    description="Enter a long paragraph, and Google's Gemini AI will summarize it!"
)

# Launch Gradio App
interface.launch(share=True)  # 'share=True' generates a public link
```

```python
# Step 1: Install Required Libraries
!pip install transformers gradio torch

# Step 2: Import Necessary Libraries
from transformers import pipeline
import gradio as gr

# Step 3: Load a Pretrained Summarization Model
summarizer = pipeline("summarization", model="t5-small")  # Alternative: use "facebook/bart-large-cnn"

# Step 4: Define the Summarization Function
def summarize_text(text):
    try:
        summary = summarizer(text, max_length=150, min_length=30, do_sample=False)
        return summary[0]['summary_text']
    except Exception as e:
        return f"Error: {str(e)}"

# Step 5: Create a Gradio Interface
interface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(label="Enter Text to Summarize"),
    outputs=gr.Textbox(label="Summarized Text"),
    title="Text Summarization with Pretrained Model",
    description="Enter a long text, and the AI model will generate a concise summary."
)

# Step 6: Launch Gradio App
interface.launch(share=True)
```
