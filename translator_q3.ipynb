```python
# ✅ Import necessary libraries
from googletrans import Translator

# ✅ Initialize Google Translate API wrapper
translator = Translator()

# ✅ Define translation function
def translate_text(text: str, target_language: str = "en") -> str:
    """
    Translates the input text to the target language using Google Translate API.
    Example: translate_text("Hola", "en") -> "Hello"
    """
    try:
        result = translator.translate(text, dest=target_language)
        return result.text
    except Exception as e:
        return f"❌ Translation failed: {str(e)}"

# ✅ Interactive loop for real-time translation
print("🌍 Multilingual Translator is ready! Type 'exit' to quit.\n")

while True:
    # Get user input
    user_input = input("Enter text to translate (or type 'exit' to quit): ").strip()

    if user_input.lower() in ["exit", "quit"]:
        print("👋 Goodbye! Happy translating!")
        break

    # Get the target language code (e.g., 'es' for Spanish, 'fr' for French, etc.)
    lang_code = input("Enter target language code (e.g., 'es' for Spanish, 'fr' for French): ").strip().lower()

    if len(lang_code) != 2:
        print("⚠️ Please enter a valid 2-letter language code like 'en', 'es', 'fr', etc.\n")
        continue

    # Translate using the function
    translation = translate_text(user_input, lang_code)
    print(f"🗣️ Translated Text: {translation}\n")
```
